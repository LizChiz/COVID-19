```r
# Creating the dataframe worldwide which separates the cases into columns based on type 
worldwide <- confirmed_cases_worldwide %>%
    pivot_wider(names_from = type, values_from = cases) 

worldwide
```

```r
worldwide1 <- worldwide %>%
    mutate(cumulative_confirmed = cumsum(confirmed))

worldwide1
```

```r
# Draw a line plot of cumulative cases vs. date, grouped and colored by is_china
# Define aesthetics within the line geom
plt_cum_confirmed_cases_china_vs_world <- ggplot(data = confirmed_cases_china_vs_world) +
  geom_line(aes(x = date, y = cum_cases, group = is_china, color = is_china)) +
  
  ylab("Cumulative confirmed cases")

# See the plot
plt_cum_confirmed_cases_china_vs_world
```

```r
# Creating an Australia dataset
Australia <- worldwide %>%
    filter(Country.Region == "Australia") %>%
    mutate(cumulative_confirmed = cumsum(confirmed)) %>%
    mutate(cumulative_death = cumsum(death)) %>%
    mutate(cumulative_recovered = cumsum(recovered))
    

Australia
```

```r
# Draw a line plot of Australias cumulative cases vs. date
# Define aesthetics within the line geom
plt_Australia <- ggplot(data = Australia, aes(x = date, y = cumulative_confirmed)) +
  geom_line() +
  ylab("Cumulative confirmed cases")

# See the plot
 plt_Australia
```
