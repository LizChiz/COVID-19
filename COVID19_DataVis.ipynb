```r
# Installing and loading tidyverse

library("tidyverse")
```

```r
# Read coronavirus_dataset.csv into cases_worldwide
cases_worldwide <- read_csv("coronavirus_dataset.csv")

# See the result
cases_worldwide
```

```r
glimpse(cases_worldwide)
```

```r
dim(cases_worldwide)
```

```r
confirmed_cases_worldwide <- cases_worldwide %>%
    filter(type == "confirmed")

dim(confirmed_cases_worldwide)
```

```r
confirmed_cases_worldwide1 <- confirmed_cases_worldwide %>%
    group_by(Country.Region) %>%
    group_by(date) %>%
    mutate(total_daily_cases = sum(cases))


dim(confirmed_cases_worldwide1) 
confirmed_cases_worldwide1
```

```r
confirmed_cases_worldwide3 <- confirmed_cases_worldwide2 %>%
    mutate(summarize(total_daily_cases = sum(cases)))


dim(confirmed_cases_worldwide3)
```

```r
# Creating an Australia dataset
Australia <- confirmed_cases_worldwide %>%
    filter(Country.Region == "Australia") %>%
    group_by(date) %>%
    summarize(total_daily_cases = sum(cases)) %>%
    mutate(cumulative_confirmed = cumsum(total_daily_cases))


Australia
```

```r
dim(Australia)
```

```r
# dropping unnecessary columns

drops <- c("Lat", "Long")
Australia <- Australia[ , !(names(Australia) %in% drops)]

dim(Australia)
```

```r
Australia1 <- Australia %>%
    group_by(date) %>%
    summarize(total_cases = sum(cases)) %>%
    mutate(cumulative_confirmed = cumsum(total_cases))

dim(Australia1)
Australia1
```

```r
# Draw a line plot of Australias cumulative cases vs. date
# Define aesthetics within the line geom
plt_Australia <- ggplot(data = Australia1, aes(x = date, y = cumulative_confirmed)) +
  geom_line() +
  ylab("Cumulative confirmed cases")

# See the plot
 plt_Australia
```
