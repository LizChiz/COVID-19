```r
# Installing and loading tidyverse

library("tidyverse")
```

```r
# Dataset cloned from this repo https://github.com/RamiKrispin/coronavirus-csv

# Read coronavirus_dataset.csv into cases_worldwide
cases_worldwide <- read_csv("coronavirus_dataset.csv")

# See the result
cases_worldwide
```

```r
glimpse(cases_worldwide)
```

```r
dim(cases_worldwide)
```

```r

length(unique(cases_worldwide$Country.Region))
unique(cases_worldwide$Country.Region)
```

```r
confirmed_cases_worldwide <- cases_worldwide %>%
    filter(type == "confirmed")

dim(confirmed_cases_worldwide)
```

```r


top_7_countries <- c("China", "US", "Italy", "Korea,South", "Iran", "Spain", "Germany")

top_7 <- confirmed_cases_worldwide %>%
    filter(Country.Region %in% top_7_countries)

top_7
```

```r
# Creating an Australia dataset
Australia <- confirmed_cases_worldwide %>%
    filter(Country.Region == "Australia") %>%
    group_by(date) %>%
    summarize(total_daily_cases = sum(cases)) %>%
    mutate(cumulative_confirmed = cumsum(total_daily_cases)) %>%
    mutate(Country.Region = "Australia")

dim(Australia)
Australia
```

```r
# joining Australia, China, Germany, Iran, Italy, South_Korea, Spain, US dataframes
x <- rbind(Australia, China, Germany, Iran, Italy, South_Korea, Spain, US)

dim(x)
x
```

```r
# Draw a line plot of x cumulative cases vs. date
# Define aesthetics within the line geom
x_plot <- ggplot(data = x, aes(x = date, y = cumulative_confirmed, group = Country.Region, color = Country.Region)) +
  geom_line() +
  ylab("Cumulative confirmed cases")

# See the plot
 x_plot
```
